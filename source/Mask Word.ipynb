```python
import pandas as pd
import matplotlib.pyplot as plt
from pythainlp import word_tokenize
import re
from pythainlp.corpus.common import thai_words
```

```python
from jiwer import wer
from ssg import syllable_tokenize

# test for one by one text file
def thai_wer(original, hypothesis):
    # delete white space
    original = re.sub(r'\s+', '', str(original))
    hypothesis = re.sub(r'\s+', '', str(hypothesis))
    # ssg engine syllable_tokenize
    ori = syllable_tokenize(original)
    hyp = syllable_tokenize(hypothesis)
    # change list to text with whitespace
    ori = re.sub(r'[\'|\'|,|\[|\]]','',str(ori))
    hyp = re.sub(r'[\'|\'|,|\[|\]]','',str(hyp))
    
    return wer(ori, hyp)

def thai_cer(original, hypothesis):
    # delete white space
    original = re.sub(r'\s+', '', str(original))
    hypothesis = re.sub(r'\s+', '', str(hypothesis))
    # ssg engine syllable_tokenize
    ori = list(original)
    hyp = list(hypothesis)
    # change list to text with whitespace
    ori = re.sub(r'[\'|\'|,|\[|\]]','',str(ori))
    hyp = re.sub(r'[\'|\'|,|\[|\]]','',str(hyp))
    
    return wer(ori, hyp)
```

```python
def cln_whitespace(list_text):
    new = [word for word in list_text if word != " "]
    return new
```

```python
def mask_word(a,b):
    output = ""
    for i in range(len(b)):
        if i < 7:
            if b[i] in a[0:i+7]:
                output += b[i]
            else:
                output +="<mask>"+b[i]+"</mask>"  
        elif i>len(a):
            if b[i] in a[-6:]:
                output += b[i]
            else:
                output +="<mask>"+b[i]+"</mask>"
        else:
            if b[i] in a[i-6:i+7]:
                output += b[i]
            else:
                output +="<mask>"+b[i]+"</mask>"
    output = output.replace('</mask><mask>','')
    return output
```
